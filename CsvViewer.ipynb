```python
from google.colab import files
import csv

def csv_ke_tabel(nama_file):
    # Membaca data dari file CSV
    with open(nama_file, newline='') as csvfile:
        pembaca = csv.reader(csvfile)
        header_asli = next(pembaca)  # Membaca header
        data = list(pembaca)

    # Menambahkan kolom No ke header dan data
    header = ['No'] + header_asli
    data_bernomer = []
    for i, baris in enumerate(data, 1):
        data_bernomer.append([str(i)] + baris)

    # Menghitung lebar maksimum untuk setiap kolom
    semua_baris = [header] + data_bernomer
    lebar = [0] * len(header)
    for baris in semua_baris:
        for i, sel in enumerate(baris):
            lebar[i] = max(lebar[i], len(sel))

    # Membuat garis pemisah
    contoh_baris = "|" + "|".join(f" {sel.ljust(lebar[i])} " for i, sel in enumerate(header)) + "|"
    pemisah = '-' * len(contoh_baris)

    # Membangun output
    hasil = []
    hasil.append(pemisah)

    # Header
    baris_header = "|" + "|".join(f" {sel.ljust(lebar[i])} " for i, sel in enumerate(header)) + "|"
    hasil.append(baris_header)
    hasil.append(pemisah)

    # Data
    for baris in data_bernomer:
        baris_str = "|" + "|".join(f" {sel.ljust(lebar[i])} " for i, sel in enumerate(baris)) + "|"
        hasil.append(baris_str)
        hasil.append(pemisah)

    return '\n'.join(hasil)

# Proses di Google Colab
uploaded = files.upload()

# Ambil nama file yang diupload
nama_file = list(uploaded.keys())[0]

# Tampilkan hasil
print("\nHasil Konversi:\n")
print(csv_ke_tabel(nama_file))
```
